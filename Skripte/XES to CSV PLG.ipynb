```python
import pandas as pd
import numpy as np 
import pm4py
from pm4py.objects.log.importer.xes import importer as xes_importer
from pm4py.objects.conversion.log import converter as log_converter
# name of xes and csv
name = "wrong_logPLG2"
# import xes 
log = xes_importer.apply(name + '.xes')
events = pm4py.get_event_attribute_values(log, 'concept:name')

# convert xes to csv
dataframe = log_converter.apply(log, variant=log_converter.Variants.TO_DATA_FRAME)

#dataframe = pd.read_csv("decline.csv", delimiter = ';')  
```

```python
#show dataframe //test
dataframe
```

```python
# create new column
dataframe['pName'] = dataframe['case:concept:name']

# create dictionary Trace <=> Name
name_di = ["Peter", "Hans", "Julian", "Kevin"]
di = {}
pi = {}  
cn = {}

for x in events:
    if "/ " in x:
        t1, t2 = x.split("/ ")
        cn[x] = t1 + "/\n" + t2
        
cn ["await answer for interview"]= "await answer\nfor interview"

for i in range(0, len(log)+1):
    x = np.random.randint(99)
    if x <=25:
        di["case_"+ str(i)] = name_di[0]
    elif x <= 50:
        di["case_"+ str(i)] = name_di[1]
    elif x <= 75:
        di["case_"+ str(i)] = name_di[2]
    else :
        di["case_"+ str(i)] = name_di[3]

        


for i in range(0, len(log)+1):
    pi["case_"+ str(i)] = "Trace "+ str(4000+i)
```

```python
# replace 
dataframe["case:concept:name"] = dataframe["case:concept:name"].replace(pi)
dataframe["concept:name"] = dataframe["concept:name"].replace(cn)
dataframe['pName'] = dataframe['pName'].replace(di)
dataframe.head(5)
```
